```python
import os, sys, io
import numpy as np
import pandas as pd
import mlflow
import mlflow.sklearn
import dagshub
import pickle

DAGSHUB_TOKEN = "02c5a3dd1ad5ef2534d5ace2e4b8b1e65f56ee46"
os.environ['MLFLOW_TRACKING_USERNAME'] = 'ashos22'
os.environ['MLFLOW_TRACKING_PASSWORD'] = DAGSHUB_TOKEN

sys.stdout = io.StringIO()
dagshub.init(repo_owner='ashos22', repo_name='House-Prices', mlflow=True)
sys.stdout = sys.__stdout__

mlflow.set_experiment('house-prices')
print('Connected to DagsHub MLflow.')
```

```python
with open('data/X_test_final_sc.pkl', 'rb') as f:
    X_test = pickle.load(f)
with open('data/test_ids.pkl', 'rb') as f:
    test_ids = pickle.load(f)

print(f'Test shape: {X_test.shape}')
print(f'Test IDs:   {len(test_ids)}')
```

```python
model = mlflow.sklearn.load_model('models:/HousePricesModel/latest')
print(f'Loaded model: {type(model).__name__}')
```

```python
log_preds  = model.predict(X_test)
predictions = np.expm1(log_preds)

submission = pd.DataFrame({'Id': test_ids, 'SalePrice': predictions})
submission.to_csv('submission.csv', index=False)

print(f'Predictions: {len(submission)}')
print(submission.head())
print('submission.csv saved.')
```
